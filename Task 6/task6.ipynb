```python
# Problem Statement 1:Smart Parking Lot Management System

def smart_parking(capacity, logs):
    current = 0
    peak = 0

    for log in logs:
        if log == "IN":
            current += 1
        elif log == "OUT":
            if current > 0:
                current -= 1

        if current > peak:
            peak = current

    print("Currently Parked Vehicles:", current)

    if current > capacity:
        print("Parking Status: FULL / OVER CAPACITY")
    else:
        print("Parking Status: Available")

    print("Peak Parking Usage:", peak)


# ---- taking input ----
capacity = int(input("Enter Parking Capacity: "))

logs_input = input('Enter Vehicle Logs (comma separated like IN,IN,OUT): ')
logs = logs_input.split(",")

smart_parking(capacity, logs)
```

```python
#  Problem Statement 2:Online Food Delivery Time Estimator

def estimate_delivery_time(distance, traffic, weather):
    base_speed = 30  # km per hour
    
    # base travel time in minutes
    time_minutes = (distance / base_speed) * 60

    # traffic delay
    if traffic.lower() == "low":
        time_minutes += 5
    elif traffic.lower() == "medium":
        time_minutes += 10
    elif traffic.lower() == "high":
        time_minutes += 20

    # weather delay
    if weather.lower() == "rainy":
        time_minutes += 15
    elif weather.lower() == "storm":
        time_minutes += 25

    print("Estimated Delivery Time:", int(time_minutes), "minutes")


# ---- taking input ----
distance = float(input("Distance (km): "))
traffic = input("Traffic Level (Low / Medium / High): ")
weather = input("Weather (Clear / Rainy / Storm): ")

estimate_delivery_time(distance, traffic, weather)


```

```python
#  Problem Statement 3:Movie Theatre Seat Occupancy Analyzer

def analyze_booking(total_seats, booked_list):
    booked_count = len(booked_list)
    
    occupancy = (booked_count / total_seats) * 100
    print("Occupancy:", int(occupancy), "%")

    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 70:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Available")

    if occupancy >= 90:
        print("Suggestion: Consider opening additional shows")


# ---- taking input ----
total_seats = int(input("Total Seats: "))

booked_input = input("Enter booked seats as 1,1,1... : ")
booked_list = booked_input.split(",")

analyze_booking(total_seats, booked_list)
```

```python
#  Problem Statement 4:Cloud Server Load Classification System

def classify_server_load(cpu_readings):
    total = 0
    
    for value in cpu_readings:
        total += value
    
    avg = total / len(cpu_readings)
    print("Average CPU Load:", int(avg), "%")
    
    if avg < 50:
        print("Server Status: Normal")
    elif avg <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


# ---- taking input ----
readings_input = input("CPU Readings (%): ")
cpu_readings = list(map(float, readings_input.split(",")))

classify_server_load(cpu_readings)
```

```python
# Problem Statement 5:Smart Classroom Resource Usage Monitor

def classroom_energy_monitor(usage_dict):
    limits = {
        "Projector": 5,
        "AC": 8,
        "Lights": 6
    }

    overused = []

    for resource in usage_dict:
        if resource in limits and usage_dict[resource] > limits[resource]:
            overused.append(resource)

    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# ---- taking input ----
usage = {}

usage["Projector"] = float(input("Projector usage (hours): "))
usage["AC"] = float(input("AC usage (hours): "))
usage["Lights"] = float(input("Lights usage (hours): "))

classroom_energy_monitor(usage)
```

```python
#  Problem Statement 6: Online Event Registration Capacity Controller

def manage_registrations(capacity, total_registrations):
    confirmed = 0
    waitlisted = 0

    for i in range(total_registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)

    if confirmed >= capacity:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")


# ---- taking input ----
capacity = int(input("Event Capacity: "))
registrations = int(input("Total Registrations: "))

manage_registrations(capacity, registrations)
```
