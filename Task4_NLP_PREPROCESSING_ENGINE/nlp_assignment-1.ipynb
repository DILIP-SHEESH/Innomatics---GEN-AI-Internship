Task 1 - Conceptual Understandig


1 . Difference between "Love" and "love" in NLP:
- To a computer, "Love" (capital 'L') and "love" (lowercase 'l') are distinct tokens with entirely different values. If not converted to lowercase, an NLP model will treat them as two separate words, splitting the frequency count and reducing the model's ability to recognize that they hold the same meaning.

2, What happens if stopwords are not removed?
- Stopwords (like "is", "the", "and") appear very often but carry very little underlying meaning. If left in, they add noise to the dataset i think, increase the dimensionality of the data (making processing slower and more expensive), and can cause models to focus on these filler words rather than the keywords that actually define the text's context.

3. Two real-world scenarios where removing stopwords can be harmful:

- Sentiment Analysis: Removing words like "not" or "no" can completely flip the meaning of a sentence (e.g., "I ain;t not happy" becomes "I ain;t happy").

- Machine Translation or Text Generation: Stopwords are essential for grammatical structure. Removing them makes it impossible to translate a sentence accurately or generate natural-sounding conversational AI text.

4. Stemming vs. Lemmatization:

- Stemming is a something, rule-based approach that chops off the ends of words to find the root (e.g., "running" becomes "run", but "better" might become "bett"). It is fast but often results in non-actual words.
- Lemmatization is a dictionary-based approach that uses morphological analysis to return the proper base form of a word, known as the lemma (e.g., "running" becomes "run", and "better" becomes "good"). It is slower but highly accurate.

```python
# Task 2: Preprocessing Function

import re

def preprocess_text(text):
    try:
        if not isinstance(text, str) or text.strip() == "":
            return [], ""
        text = text.lower()
        text = re.sub(r"http\S+|www\S+|https\S+|\S+@\S+", "", text)
        text = re.sub(r"\d+", "", text)
        text = re.sub(r'(.)\1{2,}', r'\1', text)
        text = re.sub(r'[^a-zA-Z\s]', '', text)
        text = re.sub(r'\s+', ' ', text).strip()
        tokens = text.split()
        tokens = [word for word in tokens if len(word) > 2 or word in ['no', 'not']]
        clean_sentence = " ".join(tokens)
        return tokens, clean_sentence
    except Exception as e:
        return [], ""
```

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service eveahh...0/10",
    "Call me at 98710",
    "This is THE best course this year!!!",
    "Visit https://openai.com but grok is better now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it, wait i didnt",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy/be unhappy with this"
]

for sentence in test_sentences:
    tokens, clean = preprocess_text(sentence)
    print("\nOriginal:", sentence)
    print("Tokens:", tokens)
    print("Clean Sentence:", clean)
```

```python
# Task 4: Token Analytics

def token_analysis(tokens):
    if len(tokens) == 0:
        return 0, 0, 0

    total = len(tokens)
    unique = len(set(tokens))
    avg_len = sum(len(word) for word in tokens) / total

    return total, unique, round(avg_len, 2)

for sentence in test_sentences:
    tokens, _ = preprocess_text(sentence)
    total, unique, avg = token_analysis(tokens)
    
    print("\nSentence:", sentence)
    print("Total:", total, "Unique:", unique, "Avg Length:", avg)
```

```python
# Task 5: Frequency Analysis

from collections import Counter

all_tokens = []

for sentence in test_sentences:
    tokens, _ = preprocess_text(sentence)
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 words:", counter.most_common(10))
print("Least 5 words:", counter.most_common()[-5:])
```

```python
# Task 6: Full Pipeline

def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
edge_cases = ["", "😂😂😂😂", "123456"]

for case in edge_cases:
    tokens, clean = preprocess_text(case)
    print("\nInput:", case)
    print("Tokens:", tokens)
    print("Clean:", clean)
```
