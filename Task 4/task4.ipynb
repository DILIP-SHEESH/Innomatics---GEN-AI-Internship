```python
# Problem 1: Mobile Recharge Validation System

VALID_PLANS = [199, 299, 399, 599]

def validate_recharge(amount):
    # Rule 1: Minimum recharge amount
    if amount < 50:
        print("Recharge amount must be at least ₹50.")
        return False
    
    # Rule 2: Must match a valid plan
    if amount not in VALID_PLANS:
        print("Invalid plan. Available plans:", VALID_PLANS)
        return False
    
    return True


# retry madi
while True:
    try:
        amount = int(input("Enter recharge amount: ₹"))
        
        if validate_recharge(amount):
            print("Recharge successful ✅")
            break
        else:
            print("Please try again.\n")
    
    except ValueError:
        print("Invalid input. Enter a numeric value.\n")
```

```python
# Problem 2: Inventory Reorder Alert System

inventory = {
    "Laptop": 20,
    "Mouse": 12,
    "Keyboard": 18,
    "Monitor": 10
}

def check_inventory(stock_data):
    for product, quantity in stock_data.items():
        if quantity < 15:
            print(f"{product}: Reorder Alert (Stock = {quantity})")
        else:
            print(f"{product}: Stock OK (Stock = {quantity})")

# Run the system
check_inventory(inventory)
```

```python
# Problem 3: Student Result Processing System

def process_result(marks):
    total = 0
    
    for m in marks:
        total += m
    
    average = total / len(marks)
    
    if average >= 50:
        return "Pass", average
    else:
        return "Fail", average


# Example lelete he
student_marks = [60, 45, 55, 70]

result, avg = process_result(student_marks)

print("Average Marks:", avg)
print("Result:", result)
```

```python
# Problem 4: Cab Fare Estimator with Retry Option

def fare_calc(km, peak):
    fare = 50 + (12 * km)
    if peak == "yes":
        fare = fare * 1.25
    return fare


while True:
    try:
        km = float(input("Distance in km: "))
        peak = input("Peak hour? (yes/no): ").lower()
        
        total = fare_calc(km, peak)
        print("Total Fare: ₹", total)
        break
    
    except:
        print("Invalid input, try again\n")
```

```python
# Problem 5: Employee Attendance Eligibility Checker

def check_attendance(att_list):
    present = 0
    
    for status in att_list:
        if status == "P":
            present += 1
    
    percentage = (present / len(att_list)) * 100
    
    if percentage >= 75:
        return "Eligible", percentage
    else:
        return "Not Eligible", percentage


# example klelete he 
attendance = ["P", "A", "P", "P", "A", "P", "P", "P"]

result, percent = check_attendance(attendance)

print("Attendance %:", percent)
print("Status:", result)
```

```python
# Problem 6: Password Strength Checker

def check_password(pwd):
    if len(pwd) < 8:
        return "Weak Password"
    
    has_digit = False
    has_special = False
    
    for ch in pwd:
        if ch.isdigit():
            has_digit = True
        if ch in "@#$":
            has_special = True
    
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"


# example
password = input("Enter password: ")
print(check_password(password))
```
