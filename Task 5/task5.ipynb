```python
# Problem 1: Social Media – Post Engagement Analyzer

def analyze_post(likes_list):
    total_likes = 0
    
    # total likes calculation
    for likes in likes_list:
        total_likes += likes
    
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    
    print("Total Likes:", total_likes)
    print("Post Status:", status)


# Example usage
likes = [200, 150, 300, 400]
analyze_post(likes)
```

```python
# Problem 3: Agriculture – Rainfall Adequacy Checker



def analyze_rainfall(rainfall_list, required_level):
    total = 0
    
    # total rainfall calculation
    for rain in rainfall_list:
        total += rain
    
    average = total / len(rainfall_list)
    
    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    
    print("Average Rainfall:", int(average))
    print("Rainfall Status:", status)


# Example usage
rainfall = [60, 80, 70, 78]
required = 70
analyze_rainfall(rainfall, required)
```

```python
# Problem 4: Social Media – Duplicate Account Detection

def detect_duplicates(usernames):
    unique_users = set(usernames)
    
    # simple check condition
    if len(unique_users) != len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


# Example usage
users = ["raj123", "neha07", "raj123", "amit45"]
detect_duplicates(users)
```

```python
# Problem 6: Agriculture – Premium Crop Price Filter

def filter_premium(prices):
    premium = []
    
    # loop in prices and check if greater than 2000
    
    for price in prices:
        if price > 2000:
            premium.append(price)
    
    print("Premium Crops:", premium)


# Example usage
crop_prices = [1500, 2500, 1800, 3200]
filter_premium(crop_prices)
```

```python
# Problem 7: System Monitoring – Application Health Checker

def check_system_health(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    
    print("Error Count:", errors)
    print("System Status:", status)


# Example usage
check_system_health(7)
```

```python
# Problem 8: Banking – Daily Transaction Limit Checker

def validate_transaction(amount):
    limit = 50000
    
    if amount <= limit:
        status = "Approved"
    else:
        status = "Rejected"
    
    print("Transaction Amount:", amount)
    print("Transaction Status:", status)


# Example usage
validate_transaction(60000)
```

```python
# Problem 9: E-Learning – Student Attendance Eligibility System

def check_exam_eligibility(attendance_list):
    total_days = len(attendance_list)
    present_days = 0


    # lets take present =1 and absent =0
    for day in attendance_list:
        if day == 1:     
            present_days += 1

    percentage = (present_days / total_days) * 100

    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", status)


# Example usage
attendance = [1, 1, 0, 1, 1] 
check_exam_eligibility(attendance)
```

```python
# Problem 10: Smart Electricity Bill Analyzer

def calculate_bill(units):
    bill = 0

    # Billing calculatiob based on unit consume
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    # Usage status madi 
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    return bill, status


# Example usage
units = 250
total_bill, usage_status = calculate_bill(units)

print("Units Consumed:", units)
print("Total Bill:", total_bill)
print("Usage Level:", usage_status)
```
